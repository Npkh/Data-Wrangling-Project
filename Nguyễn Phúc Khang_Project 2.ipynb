# Use any data wrangling tools to calculate the total revenue of each product at each price. 

## Cách 1: Sử dụng tách bảng và merge theo điều kiện

```python
import pandas as pd

# Import hai file csv
sale = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\sales.csv")
price = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\prices.csv")

#Format lại thời gian
sale['ordered_at'] = pd.to_datetime(sale['ordered_at'])
price['updated_at'] = pd.to_datetime(price['updated_at'])

#Sort value theo thời gian
sale = sale.sort_values(by = 'ordered_at')
price = price.sort_values(by = 'updated_at')

# Gộp hai file vào với nhau, cách gộp: so sánh ordered_at ở cột trái và updated_at ở cột phải với nhau
sale_report = pd.merge_asof(sale, price, left_on = 'ordered_at', right_on = 'updated_at', by = 'product_id', direction = 'nearest')
sale_report
```

```python
#Tạo ra một bảng phụ mà thời gian đặt hàng (ordered_at) < thời gian cập nhật giá (updated_at)
old_sale_report = sale_report[sale_report['ordered_at'] < sale_report['updated_at']].copy()

#Tạo ra cột doanh thu mới bằng cách lấy số lượng bán ra (quantity_ordered) * giá cũ (old_price)
old_sale_report['revenue'] = old_sale_report['quantity_ordered'] * old_sale_report['old_price']
old_sale_report
```

```python
#Tạo ra một bảng phụ mà thời gian đặt hàng (ordered_at) >= thời gian cập nhật giá (updated_at)
new_sale_report = sale_report[sale_report['ordered_at'] >= sale_report['updated_at']].copy()

#Tạo ra cột doanh thu mới bằng cách lấy số lượng bán ra (quantity_ordered) * giá mới (new_price)
new_sale_report['revenue'] = new_sale_report['quantity_ordered'] * new_sale_report['new_price']
new_sale_report
```

```python
#Gộp hai bảng phụ lại với nhau theo chiều dọc
total_revenue = pd.concat([old_sale_report, new_sale_report])
total_revenue
```

```python
#Tính tổng doanh thu theo product_id
result1 = total_revenue.groupby('product_id').agg({'revenue': 'sum'})
result1
```

## Cách 2: Sử dụng Lambda

```python
import pandas as pd

# Import hai file csv
sale = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\sales.csv")
price = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\prices.csv")

#Format lại thời gian
sale['ordered_at'] = pd.to_datetime(sale['ordered_at'])
price['updated_at'] = pd.to_datetime(price['updated_at'])

#Sort value theo thời gian
sale = sale.sort_values(by = 'ordered_at')
price = price.sort_values(by = 'updated_at')

# Gộp hai file vào với nhau
sale_report = pd.merge_asof(sale, price, left_on = 'ordered_at', right_on = 'updated_at', by = 'product_id', direction = 'nearest')
sale_report

# Tính tổng doanh thu cho từng sản phẩm theo từng hàng
# Tại đây cần sử dụng sale_report.apply() để áp dụng lambda cho từng hàng trong dataframe sale_report
# axis = 1 nghĩa là đang tính toán theo chiều ngang (hàng), axis = 0 nghĩa là tính toán theo chiều dọc (cột)
sale_report['revenue'] = sale_report.apply(lambda x: x['quantity_ordered'] * x['old_price'] 
                                            if x['ordered_at'] < x['updated_at'] 
                                            else x['quantity_ordered'] * x['new_price'], axis=1)

# Group by product_id and sum revenue
result2 = sale_report.groupby('product_id').agg({'revenue':'sum'})
result2
```

## Cách 3: Sử dụng def

```python
import pandas as pd

#Xây dựng một hàm mới tên là total_revenue với tham số là x, x ở đây có thể hiểu là giá trị của từng hàng
def total_revenue(x):
    if x['ordered_at'] < x['updated_at']: # Nếu giá trị x ở cột ordered_at < giá trị x ở cột updated_at
        return x['quantity_ordered'] * x['old_price']
    else:
        return x['quantity_ordered'] * x['new_price']

# Import hai file csv
sale = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\sales.csv")
price = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\prices.csv")

#Format lại thời gian
sale['ordered_at'] = pd.to_datetime(sale['ordered_at'])
price['updated_at'] = pd.to_datetime(price['updated_at'])

#Sort value theo thời gian
sale = sale.sort_values(by = 'ordered_at')
price = price.sort_values(by = 'updated_at')

# Gộp hai file vào với nhau
sale_report = pd.merge_asof(sale, price, left_on = 'ordered_at', right_on = 'updated_at', by = 'product_id', direction = 'nearest')
sale_report

# Tính tổng doanh thu cho từng sản phẩm theo từng hàng
sale_report['revenue'] = sale_report.apply(total_revenue, axis=1)

# Group by product_id and sum revenue
result3 = sale_report.groupby('product_id').agg({'revenue':'sum'})
result3
```

## Cách 4: Sử dụng loc()

```python
import pandas as pd

# Import hai file csv
sale = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\sales.csv")
price = pd.read_csv("E:\\Data Analyst\\Python\\project 2\\Data Wrangling Practise\\prices.csv")

#Format lại thời gian
sale['ordered_at'] = pd.to_datetime(sale['ordered_at'])
price['updated_at'] = pd.to_datetime(price['updated_at'])

#Sort value theo thời gian
sale = sale.sort_values(by = 'ordered_at')
price = price.sort_values(by = 'updated_at')

# Gộp hai file vào với nhau
sale_report = pd.merge_asof(sale, price, left_on = 'ordered_at', right_on = 'updated_at', by = 'product_id', direction = 'nearest')
sale_report

# Tính doanh thu cho từng sản phẩm
# Hàm loc() này tạo ra cột mới revenue và xét điều kiện ordered_at < updated_at nếu thỏa mãn thì quantity * old_price 
sale_report.loc[sale_report['ordered_at'] < sale_report['updated_at'], 'revenue'] = sale_report['quantity_ordered'] * sale_report['old_price']
sale_report.loc[sale_report['ordered_at'] >= sale_report['updated_at'], 'revenue'] = sale_report['quantity_ordered'] * sale_report['new_price']

# Tổng doanh thu theo sản phẩm
result4 = sale_report.groupby('product_id').agg({'revenue':'sum'})
result4
```
